```python
"""
copyright: abhirup.ghosh.184098@gmail.com
"""

# dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize
from matplotlib.gridspec import GridSpec
import geopandas as gpd

import matplotlib as mpl
import seaborn as sns

import matplotlib.image as mpimg
import imageio

import os
```

```python
kol = gpd.read_file('../data/KOL_WARDS/kolkata_wards.shp')
ber = gpd.read_file('../data/berlin/plz.geojson')
```

```python
kol_center = (kol.centroid.x.mean(), kol.centroid.y.mean())
ber_center = (ber.centroid.x.mean(), ber.centroid.y.mean())
```

```python
kol_trans = kol.translate(xoff=-kol_center[0], yoff=-kol_center[1]+0.02)
ber_trans = ber.translate(xoff=-ber_center[0], yoff=-ber_center[1])
```

```python
ber_trans.area.sum()/kol_trans.area.sum()
```

```python
#plt.rcParams['figure.figsize'] = [16, 11]

fig, ax = plt.subplots()
ber_trans.plot(ax=ax, color='#8d99ae')
kol_trans.plot(ax=ax, color='#faf0ca')

ax.set(aspect=1.6, 
    facecolor='#1b263b'
)

ax.spines[['left','right', 'top', 'bottom']].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])

ax.text(-0.28,-0.15, '**Credit: Abhirup Ghosh/OpenStreetMap', fontsize=10, color='w', 
                            #fontstretch='ultra-expanded', 
                            fontfamily='Arial Narrow', 
                            #fontweight='bold',
                            ha='left', alpha=1)                                                    
```
